## Part 2 — The Receipt & Loyalty Stars  *(25 points)*

A customer orders **3 lattes** ($4.50 each) and **2 muffins** ($3.25 each). Texas sales tax is **8.25%**. The shop's loyalty program gives **1 star for every whole $5 spent** (based on the total).

### What to do
1. Store the four order values in variables: `latte_price`, `latte_qty`, `muffin_price`, `muffin_qty`.
2. Calculate, using **arithmetic operators**:
   - `subtotal` — price × quantity for both items, added together
   - `tax` — `subtotal * 8.25 / 100`
   - `total` — subtotal plus tax
3. Print the receipt using **f-strings with 2 decimal places**, like this:

   ```
   Subtotal: $20.00
   Tax (8.25%): $1.65
   Total: $21.65
   ```

   > 💡 **Money formatting recipe:** `f"${subtotal:.2f}"` prints a float with exactly 2 decimals.
4. Loyalty stars: first convert the total to whole dollars with `int(total)` (store it in `total_whole`). Then use **floor division and modulo**:
   - `stars_earned = total_whole // 5`
   - `toward_next = total_whole % 5`
   
   Print both with labels, e.g. `Stars earned: 4` and `Dollars toward next star: 1`.
5. The customer brought a **$25.00 budget**. Compute and print how much money they have left, formatted as money.
6. Add a **comment** answering: why do `//` and `%` together "split" the dollars into stars and leftover? (One sentence — hint: `21 = 4 * 5 + 1`.)


````python
# Bida Sakya Part 2

#1
latte_price = 4.50
latte_qty = 3
muffin_price = 3.25
muffin_qty = 2

#2 arithmetic operation
subtotal = (latte_price * latte_qty) + (muffin_price * muffin_qty)
tax = subtotal * 8.25 / 100
total = subtotal + tax

#3
print(f"Subtotal: ${subtotal:.2f}")
print(f"Tax(8.25%):${tax:.2f}")
print(f"Total:${total:.2f}")

#4
total_whole =int(total)
stars_earned = total_whole // 5
toward_next = total_whole % 5
print(f"Stars_earned:{stars_earned}")
print(f"Dollars toward next star:{toward_next}")

#5
budget= 25
money_left = budget - total_whole
print(f"Money left: ${money_left}")

#6
# // gives the number of complete groups of 5 dollars, and %
# gives the leftover dollars after making those groups.

````

## Part 5 — The Weekly Prize Drawing  *(20 points)*

Every Friday, Eagle Beans runs a prize drawing for its customers — and shuffles the café playlist.

### What to do
The starter lists are provided:
```python
customers = ["Maria", "Devon", "Aisha", "Carlos", "Lena", "Tyler"]
songs = ["Jazz Vibes", "Lo-Fi Beats", "Morning Acoustic", "Indie Mix", "Soft Piano"]
```
1. Import the `random` module and set a **seed** (any whole number) so the drawing is reproducible.
2. Print, with labels:
   - this week's **lucky number**: a random integer from 1 to 50 (`random.randint`)
   - a **discount roll**: a random float between 0 and 1 (`random.random`)
3. The drawing:
   - **one** free-drink winner with `random.choice(...)`
   - **two different** coupon winners with `random.sample(..., 2)`
4. Shuffle the `songs` list with `random.shuffle(...)` and print the new playlist order.
5. Add a **comment**: run the cell twice — do the winners change? What would happen **without** the seed line? (One or two sentences.)

### Hints
- `random.shuffle(songs)` changes the list **in place** and returns `None` — shuffle first, print the list on the next line.


````python
customers = ["Maria", "Devon", "Aisha", "Carlos", "Lena", "Tyler"]
songs = ["Jazz Vibes", "Lo-Fi Beats", "Morning Acoustic", "Indie Mix", "Soft Piano"]

# Bida Shakya Part 5
#1
import random
random.seed(54)

#2
print("This week's lucky number:", random.randint(1,50))
print("A discount roll:", random.random())

#3
print("One free drinker:", random.choice(customers))
print("Two different coupon winners:", random.sample(customers,2))

#4
random.shuffle(songs)
print("The new playlist order:",songs )

#5
#Since random.seed(54) is used, running the cell multiple times produces the same "random" results.
# Without the seed line, the lucky number, winners, and shuffled playlist would likely be different each time.
````
